```python
df["Make"] = df["Model"].str.split().str[0]

print(df["Make"].nunique())
```

```python
df["Model_No_Year"] = df["Model"].str.replace(
    r"^\d{4}\s+",
    "",
    regex=True
)
```

```python
df["Make"] = df["Model_No_Year"].str.split().str[0]
```

```python
print(df["Make"].value_counts().head(20))
```

```python
import numpy as np

df["log_price"] = np.log1p(df["Price"])
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,4))
plt.hist(df["log_price"], bins=50)
plt.title("Log Price Distribution")
plt.xlabel("log_price")
plt.ylabel("Count")
plt.show()
```
